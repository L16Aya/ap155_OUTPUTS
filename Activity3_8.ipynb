```python
import numpy as np
from numpy import logspace
from numpy import cos, sin
import matplotlib.pyplot as plt

def func(x):
  return (1.0-cos(x))/x**2.0

def fpp(x):
    return (6.0-4.0*x*sin(x)+((x**2)-6)*cos(x))/x**4.0

def secdev(func, x, dx):
  return (func(x-dx)-2.0*func(x)+func(x+dx))/dx**2.0

dx = logspace(-1, -6, 120)

absdiff = [secdev(func, 0.004, dxi) - fpp(0.004) for dxi in dx]

plt.loglog(dx, np.abs(absdiff), "o")
plt.xlabel("dx")
plt.ylabel("Absolute Error")
plt.title("Error of Second Derivative Approximation")
plt.show()





```

```python
import numpy as np
from numpy import logspace
from numpy import cos, sin
import matplotlib.pyplot as plt

def func(x):
  return (1.0-cos(x))/x**2.0

def g(x):
  return (2*(sin(x/2))**2)/x**2.0

def fpp(x):
    return (6.0-4.0*x*sin(x)+((x**2)-6)*cos(x))/x**4.0

def secdev(func, x, dx):
  return (func(x-dx)-2.0*func(x)+func(x+dx))/dx**2.0

def secdev2(func, x, dx):
  return (g(x-dx)-2.0*g(x)+g(x+dx))/dx**2.0

dx = logspace(-1, -6, 120)

absdiff = [secdev(func, 0.004, dxi) - fpp(0.004) for dxi in dx]
absdiff2 = [secdev2(g, 0.004, dxi) - fpp(0.004) for dxi in dx]

plt.loglog(dx, np.abs(absdiff), "o")
plt.loglog(dx, np.abs(absdiff2), "o")
plt.xlabel("dx")
plt.ylabel("Absolute Error")
plt.title("Error of Second Derivative Approximation")
plt.show()





```
