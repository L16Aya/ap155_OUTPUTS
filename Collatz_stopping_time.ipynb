```python
import matplotlib.pyplot as plt

def collatz_steps(N):
  i = 0
  while N != 1:
    if N % 2 == 0:
        N /= 2
    else:
        N = (3 * N) + 1
    i+=1
  return i

Ns=[]
Tn=[]

for N in range(1,10001):
  Ns.append(N)
  Tn.append(collatz_steps(N))

plt.scatter(Ns,Tn,s=10)
plt.xlabel("N")
plt.ylabel("T(N)")
plt.show()



```
