```python
import numpy as np
import math
import matplotlib.pyplot as plt
from math import sqrt

def function(x):
  return -x**5 + 4*x**4 - 4*x**3 + (x**2)*(np.exp(x)) - 2*x**2 - (4*x)*(np.exp(x)) + 8*x + 4*np.exp(x) - 8

def ffd(x):
  return -5*x**4 + 16*x**3 - 12*x**2 + (x**2)*(np.exp(x)) + (2*x)*(np.exp(x)) - 4*x - (4*x)*(np.exp(x)) - 4*(np.exp(x)) + 8 + 4*np.exp(x)

def fsd(x):
  return -20*x**3 + 48*x**2 - 24*x + (x**2)*(np.exp(x)) + (2*x)*(np.exp(x)) + 2*x*(np.exp(x)) + 2*(np.exp(x)) - 4 - (4*x)*(np.exp(x)) - 4*(np.exp(x))

def phi(x):
  return x**4 - 9*x**3 + 25*x**2 - 24*x + 4

def pfd(x):
  return 4*x**3 - 27*x**2 + 50*x - 24

def psd(x):
  return 12*x**2 - 54*x + 50

def golden(phi, x0, x1, kmax=200, tol=1.e-8):
  varphi = 0.5*(1 + sqrt(5))
  for k in range(1,kmax):
    x2 = x0 + (x1-x0)/(varphi+1)
    x3 = x0 + (x1-x0)*varphi/(varphi+1)

    if phi(x3) < phi(x2):
      x0 = x2
    else:
      x1 = x3

    xnew = (x0+x1)/2
    xdiff = abs(x1-x0)

    if abs(xdiff) < tol:
      break

  else:
    xnew = None
  return xnew

x = np.linspace(-100, 100, 5000)

print(golden(function,0.,5.))
plt.plot(x, function(x), label= 'f(x)')
plt.plot(x, ffd(x), label="f'(x)")
plt.plot(x, fsd(x), label='f"(x)')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Golden Section Method")
plt.text(-1, -30, "minima: " + str(golden(function,0.,5.)) , fontsize=12)
plt.ylim(-50, 10)
plt.xlim(-2, 5)
plt.legend()

```
