```python
import numpy as np
from numpy import logspace
from numpy import cos, sin
import matplotlib.pyplot as plt

def f(x):
  return 2.0 + 5.0*np.sin(x) + 0.1*np.sin(30*x)

def g(x):
  return 2.0 + 5.0*np.sin(x)

def ffdev(x):
  return 5.0*np.cos(x) + 3.0*np.cos(30*x)

def gfdev(x):
  return 5.0*np.cos(x)

def fdapp(f, x, dx):
  return (f(x+dx)-f(x))/dx

x = np.linspace(0, 2*np.pi, 128)
dx = x[1] - x[0]

plt.plot(x, ffdev(x), label="analytical f'(x)")
plt.plot(x, gfdev(x), label="analytical g'(x)")
plt.plot(x, fdapp(f, x, dx), label="forward diff (h))")
plt.xlabel("x")
plt.xlabel("y")
plt.legend()
```

```python
import numpy as np
from numpy import logspace
from numpy import cos, sin
import matplotlib.pyplot as plt

def f(x):
  return 2.0 + 5.0*np.sin(x) + 0.1*np.sin(30*x)

def g(x):
  return 2.0 + 5.0*np.sin(x)

def ffdev(x):
  return 5.0*np.cos(x) + 3.0*np.cos(30*x)

def gfdev(x):
  return 5.0*np.cos(x)

def fdapp(f, x, dx):
  return (f(x+dx)-f(x))/dx

def fdapp2(f, x, dx):
  return (f(x+2*dx)-f(x))/(2*dx)

x = np.linspace(0, 2*np.pi, 128)
dx = x[1] - x[0]

plt.plot(x, ffdev(x), label="analytical f'(x)")
plt.plot(x, gfdev(x), label="analytical g'(x)")
plt.plot(x, fdapp(f, x, dx), label="forward diff (h)")
plt.plot(x, fdapp2(f, x, dx), label="forward diff (2h)")
plt.xlabel("x")
plt.xlabel("y")
plt.legend()
```
