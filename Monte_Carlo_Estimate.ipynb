```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return ((x**2 + x) / (x**(1/2))) * np.exp(-x)

def f2(x):
    return np.exp(np.sin(2*x))

def f3(x):
    return 1 / (2 + np.cos(x))

def f4(x):
    return np.exp((-x**2) / 2)

def f5(x):
    return (x ** 0.5) * np.cos(x)

x = np.linspace(-10, 10, 400)

plt.figure(figsize=(8, 6))

plt.plot(x, f1(x), label='f1(x)', linewidth=2)
plt.plot(x, f2(x), label='f2(x)', linewidth=2)
plt.plot(x, f3(x), label='f3(x)', linewidth=2)
plt.plot(x, f4(x), label='f4(x)', linewidth=2)
plt.plot(x, f5(x), label='f5(x)', linewidth=2)

plt.title('Plot of 5 Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import integrate
from math import sqrt, exp, log

N = 100000
u = np.random.uniform(size=N)

# simplify the function to (x**1.5 + x**0.5)*e^-x
# p(x)=e^-x
# p^-1(x)=-log(u)
x = -np.log(u)
f = x**1.5 + x**0.5

I_mc = np.mean(f)

print("Monte Carlo estimate:")
print("I  " + str(I_mc))
print("")

I_exact, err = integrate.quad(f1, 0, np.inf)

print("SciPy integrate.quad result:")
print("I = " + str(I_exact))
```

```python
# p(x)=1/2pi
# p^-1(x)=2pi
x = np.random.random(N) * 2*np.pi
f = 2*np.pi * np.exp(np.sin(2*x))

I_mc = np.mean(f)

print("Monte Carlo estimate:")
print("I  " + str(I_mc))
print("")

I_exact, err = integrate.quad(f2, 0, 2*np.pi)

print("SciPy integrate.quad result:")
print("I = " + str(I_exact))
```

```python
# p(x)=1/2pi
# p^-1(x)=2pi
x = np.random.random(N) * 2*np.pi
f = 2*np.pi / (2 + np.cos(x))

I_mc = np.mean(f)

print("Monte Carlo estimate:")
print("I  " + str(I_mc))
print("")

I_exact, err = integrate.quad(f3, 0, 2*np.pi)

print("SciPy integrate.quad result:")
print("I = " + str(I_exact))
```

```python
# p(x)=1/2pi
# p^-1(x)=2pi
x = np.random.random(N) * 2*np.pi
f = 2*np.pi * np.exp((-x**2) / 2)

I_mc = np.mean(f)

print("Monte Carlo estimate:")
print("I  " + str(I_mc))
print("")

I_exact, err = integrate.quad(f4, 0, 2*np.pi)

print("SciPy integrate.quad result:")
print("I = " + str(I_exact))
```

```python
# p(x)=1/pi
# p^-1(x)=pi
x = np.random.random(N) * np.pi
f = np.pi * x**0.5 * np.cos(x)

I_mc = np.mean(f)

print("Monte Carlo estimate:")
print("I  " + str(I_mc))
print("")

I_exact, err = integrate.quad(f5, 0, np.pi)

print("SciPy integrate.quad result:")
print("I = " + str(I_exact))
```
